#### **üìò Metric Evaluation :**

To assess the performance of the classification models, we use the following evaluation metrics:
- **Accuracy**  
  Measures the overall correctness of the model by calculating the ratio of correctly predicted observations to the total observations.  
  $$
  \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
  $$
- **Precision**  
  Indicates how many of the predicted positive cases were actually positive. It is especially important when the cost of false positives is high.  
  $$
  \text{Precision} = \frac{TP}{TP + FP}
  $$
- **Recall (Sensitivity)**  
  Measures the model's ability to correctly identify actual positive cases. It is crucial when the cost of false negatives is high.  
  $$
  \text{Recall} = \frac{TP}{TP + FN}
  $$
- **F1-Score**  
  The harmonic mean of precision and recall. It provides a balance between the two, especially useful in cases of class imbalance.  
  $$
  \text{F1-Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
  $$
> üí° **Note**: In the context of this project, since the dataset is likely imbalanced (more 'no' than 'yes'), precision and recall are more informative than accuracy.


#### **üìò Why F1-Score Matters Most in This Project?**

In this marketing campaign scenario, the **F1-Score** is the most suitable evaluation metric. It strikes a balance between **precision** (avoiding wasteful outreach) and **recall** (capturing potential subscribers). This ensures that:

- The marketing team doesn't target too many uninterested customers (low FP).
- The model doesn‚Äôt miss valuable prospects (low FN).
- Campaigns remain cost-effective and revenue-driven.

This dataset contains four numerical columns: age, balance, campaign, and pdays. We will examine these columns to detect any outliers or unusual values.

```python
numerical_cols = ['age', 'balance', 'campaign', 'pdays']
```

```python
def outliers_iqr(df, columns):
    outliers = {}
    for i in columns:
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outlier_data = df[(df[i] < lower_bound) | (df[i] > upper_bound)][i]
        outliers[i] = outlier_data.index.tolist()

        print(f"{i}: {len(outlier_data)} ouliters detected")
        print(f"Lower bound: {lower_bound}, upper bound: {upper_bound}")

    return outliers

outlier_results = outliers_iqr(df, numerical_cols)
```

<h1 style='text-align:left; color:#ffc107; font-size: 24px'><b> Section 5: Modeling </b></h1>

```python
categorical_cols_onehot = ['job', 'contact', 'poutcome']
categorical_cols_ordinal = ['housing', 'loan']
month_col = ['month']
numeric_cols_balance = ['balance']
numeric_cols_other = ['age', 'campaign', 'pdays', 'ever_contacted']
```

```python
# Pipeline for categorical columns using One-Hot Encoding
onehot_pipeline = Pipeline([
    ('fill_missing', SimpleImputer(strategy='constant', fill_value='unknown')),  # Fill missing values with 'unknown'
    ('one_hot_encode', OneHotEncoder(handle_unknown='ignore', sparse_output=False))  # One-hot encoding, ignore unknown categories
])

# Pipeline for categorical columns using Binary Encoding
binary_pipeline = Pipeline([
    ('fill_missing', SimpleImputer(strategy='constant', fill_value='unknown')),  # Fill missing values with 'unknown'
    ('binary_encode', BinaryEncoder())  # Apply binary encoding to categorical features
])

# Special pipeline for the 'month' column with Ordinal Encoding based on month order
month_pipeline = Pipeline([
    ('fill_missing', SimpleImputer(strategy='most_frequent')),  # Fill missing values with the most frequent month
    ('ordinal_encode', OrdinalEncoder(categories=[['jan', 'feb', 'mar', 'apr', 'may', 'jun',
                                                   'jul', 'aug', 'sep', 'oct', 'nov', 'dec']]))  # Ordinal encoding with month order
])

# Pipeline for other categorical columns using automatic Ordinal Encoding
ordinal_pipeline = Pipeline([
    ('fill_missing', SimpleImputer(strategy='most_frequent')),  # Fill missing values with the most frequent value
    ('ordinal_encode', OrdinalEncoder(categories='auto'))  # Ordinal encoding with automatic category detection
])


negative_to_positive = FunctionTransformer(np.abs)  

# Pipeline for the 'balance' column
balance_pipeline = Pipeline([
    ('neg_to_pos', negative_to_positive),  # Convert negative values to positive
    ('fill_median', SimpleImputer(strategy='median')),  # Fill missing values with median
    ('scale_robust', RobustScaler())  # Scale using RobustScaler
])

# Pipeline for other numeric columns
numeric_pipeline = Pipeline([
    ('fill_median', SimpleImputer(strategy='median')),  # Fill missing values with median
    ('scale_robust', RobustScaler())  # Scale using RobustScaler
])
```

```python
preprocessor = ColumnTransformer([
    ('onehot', onehot_pipeline, categorical_cols_onehot),
    ('month', month_pipeline, month_col),
    ('ordinal', ordinal_pipeline, categorical_cols_ordinal),
    ('balance', balance_pipeline, numeric_cols_balance),
    ('numeric', numeric_pipeline, numeric_cols_other)
])
```

```python
# --- THRESHOLD OPTIMIZATION STAGE ---

final_champion_model = best_models['Random Forest']
y_proba_test = final_champion_model.predict_proba(X_test_fe)[:, 1]

threshold_analysis = []

thresholds = np.arange(0.1, 0.9, 0.01)

# Loop through each threshold
for th in thresholds:
    y_pred_new = [1 if prob >= th else 0 for prob in y_proba_test]

    # Calculate confusion matrix
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_new).ravel()

    # Calculate Net Benefit
    net_benefit = (tp * 1419) - (fp * 341) - (fn * 1419)

    # Calculate Precision and Recall
    precision = precision_score(y_test, y_pred_new, zero_division=0)
    recall = recall_score(y_test, y_pred_new, zero_division=0)

    # Store the results
    threshold_analysis.append({
        'Threshold': th,
        'Net_Benefit': net_benefit,
        'Precision': precision,
        'Recall': recall,
        'TP': tp,
        'FP': fp,
        'FN': fn
    })
```

```python
threshold_df = pd.DataFrame(threshold_analysis)

optimal_row = threshold_df.loc[threshold_df['Net_Benefit'].idxmax()]
optimal_threshold = optimal_row['Threshold']
```

```python
print("--- Threshold Optimization Results ---")
display(threshold_df.sort_values(by='Net_Benefit', ascending=False).head(10))

print("\n--- Optimal Threshold Found ---")
print(f"Best Threshold: {optimal_row['Threshold']:.2f}")
print(f"Maximum Net Benefit: ‚Ç¨{optimal_row['Net_Benefit']:,.2f}")
print(f"With Precision: {optimal_row['Precision']:.2f} and Recall: {optimal_row['Recall']:.2f}")

# Visualize the results
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(threshold_df['Threshold'], threshold_df['Net_Benefit'], color='red', label='Net Benefit')
ax.plot(threshold_df['Threshold'], threshold_df['Precision'], color='blue', linestyle='--', label='Precision')
ax.plot(threshold_df['Threshold'], threshold_df['Recall'], color='green', linestyle='--', label='Recall')

# Mark the optimal point
ax.axvline(x=optimal_threshold, color='purple', linestyle=':', label=f'Optimal Threshold ({optimal_threshold:.2f})')

ax.set_title('Threshold Optimization Analysis vs. Net Benefit', fontsize=16)
ax.set_xlabel('Probability Threshold')
ax.set_ylabel('Score / Value')
ax.legend()
plt.show()
```

### Threshold Optimization Chart Explanation

This chart shows how our key business metrics change when we adjust the **probability threshold** of our Random Forest model.  
The X-axis represents the tested threshold values, from more lenient (left) to stricter (right).

- **Precision (blue)** and **Recall (green)** show the classic trade-off:
  - Lower thresholds ‚Üí higher recall (we capture more potential customers) but lower precision (more incorrect 'Yes' predictions).
  - Higher thresholds ‚Üí higher precision (more accurate positives) but lower recall (we miss more potential customers).

- **Net Benefit (red)** is the most important curve here, as it reflects the financial outcome:
  - At very low thresholds, net benefit is low because false positives cost money.
  - As we increase the threshold, net benefit improves thanks to more correct positive predictions.
  - If the threshold becomes too high, net benefit drops again because we lose too many potential customers.

The **purple dashed line** at around **0.35** marks the optimal threshold ‚Äî the point where the balance between gaining new customers (recall) and controlling costs (precision) produces the highest net benefit.  
This threshold is recommended as the new decision standard for the model.
